## 글 크롤링

```python
# "url_list.csv" 불러오기
url_load = pd.read_csv('url_list.csv')        # 기본 모델
url_load
```

# 크롤링하기

```python
dict = {}  # 전체 크롤링 데이터를 담을 그릇
```

```python
for i in tqdm(range(0, len(url_load))):
    try:
        # 글 띄우기
        url = url_load['url'][i]
        driver.get(url)   # 최신글 띄우기
        time.sleep(1.5)
        
        # iframe 접근
        driver.switch_to_frame('cafe_main')

        target_info = {}   # 글 하나의 딕셔너리
    #     # 카테고리 크롤링 시작
    #     overlays = "#main-area .inbox .fl a.m-tcol-c"
    #     cate = driver.find_element_by_css_selector(overlays)      # category
    #     category = cate.text
    #     category

        # 제목 크롤링 시작
        overlays = ".title_text"                                 
        tit = driver.find_element_by_css_selector(overlays)         # title
        title = tit.text
        title
        
        # 내용(content) 크롤링 시작
        overlays = ".ContentRenderer"                                  
        cont = driver.find_element_by_css_selector(overlays)        # content
        content = cont.text
        content

        # 작성자(id) 크롤링 시작
        overlays = ".nickname"                                  
        id_ = driver.find_element_by_css_selector(overlays)         # person
        person = id_.text
        person

        # 날짜(date) 크롤링 시작
        overlays = ".date"                
        date_ = driver.find_element_by_css_selector(overlays)        # datetime
        datetime = date_.text
        datetime

        # 댓글(reply) 크롤링 시작
        overlays = ".text_comment"                                  
        reply = driver.find_elements_by_css_selector(overlays)      # re_list
        r_num = len(reply)
        re_list = []       
        for re in reply:    
            re_list.append(re.text)
        replies = ",".join(re_list)
        replies
        
#         # 이미지(image) 크롤링 시작
#         overlays = "article_img.ATTACH_IMAGE"                                 
#         img_tmp = driver.find_elements_by_name(overlays)         # image

#         img_list = []   
#         for img in img_tmp:    
#             img_url = img.get_attribute('src')
#             img_list.append(img_url)
#         image = ",".join(img_list)
#         image 

        # 글 1개 딕셔너리에 크롤링한 데이터 집어넣기
        #target_info['category'] = category
        target_info['title'] = title
        target_info['url'] = url
        target_info['content'] = content
        target_info['person'] = person
        target_info['datetime'] = datetime
        target_info['reply'] = replies
#         target_info['image'] = image

        # 글 전체 딕셔너리에 글 각각 딕셔너리 넣기
        # target_info

        dict[i] = target_info
        time.sleep(1)

        print(i, title)
    
    except:        
        continue   
```
